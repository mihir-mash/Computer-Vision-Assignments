```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
from google.colab import files

uploaded = files.upload()
filename = list(uploaded.keys())[0]
```

```python
img = cv2.imread(filename)

cv2_imshow(img)

#Print image size
h, w = img.shape[:2]
print("Image size (Height x Width):", h, "x", w)
```

```python
#Pixel values
print("Sample Pixel Values (Color BGR):")
print("Top-left pixel:", img[0,0])
print("Center pixel:", img[h//2, w//2])
print("Bottom-right pixel:", img[h-1, w-1])
```

```python
#Grayscale conversion
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

print("\nConverted to Grayscale")
cv2_imshow(gray)
```

```python
#Histogram (Grayscale)
plt.figure(figsize=(6,4))
plt.hist(gray.ravel(), bins=256, range=[0,256])
plt.title("Grayscale Histogram")
plt.xlabel("Intensity (0=dark, 255=bright)")
plt.ylabel("Number of pixels")
plt.show()
```

```python
#RGB Histogram
colors = ('b','g','r')

plt.figure(figsize=(6,4))
for i, col in enumerate(colors):
    hist = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(hist, color=col)

plt.title("RGB Histogram")
plt.xlabel("Intensity")
plt.ylabel("Count")
plt.show()
```

```python
#Show grayscale as array (NEW PART)
print("\nGrayscale image as array (first 5x5 pixels):")
print(gray[:5, :5])   # only small part for readability

print("\nGrayscale center pixel value:", gray[h//2, w//2])

print("\nMin intensity:", gray.min())
print("Max intensity:", gray.max())
```
